# Caso 3
Casos donde los nuevos sufijos no están incluídos en sufijos que se usanron en el tageo, pero que son substrings de sufijos que sí estaban. los buco en freqs teniendo en cuenta que no contengan el string entero.

```python
numero = ['singular',  'plural']
nuevos_sufijos = [('ión','iones'),
                    ('ma','mas'),
                    ('ud','udes')]

viejo_sufijos = [(('ción','sión'),('ciones','siones')),
                 (('oma','ema'),('omas','emas')),
                 (('tud','itud'),('tudes','itudes'))]


init = True
for p, i in enumerate(nuevos_sufijos):
    for j, n in enumerate(i):
        print(n, numero[j])
        m = n + '$'
        ind1 = frecs['Palabra'].str.contains(m, regex=True)
        tmp1 = frecs.loc[ind1]
        tmp1.drop('frecuencia', axis=1,inplace=True)
        tmp1['Sufijo_original'] = '-'
        tmp1['R'] = '-'
        tmp1['Sufijo_nuevo'] = i[0]
        tmp1['Número'] = numero[j]
        
        for a in viejo_sufijos[p][j]:
            a2 = a + '$'
            ind2 = tmp1['Palabra'].str.contains(a2, regex=True)
            tmp1 = tmp1[-ind2]

        if init:
            df = tmp1
            init= False
        else:
            df = df.append(tmp1)


df.to_csv('Caso3.csv',index=False)
```
